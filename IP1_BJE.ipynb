```python
#4.)
car_df=pd.read_csv('car-sales.csv')
car_df
```

```python
#5.)
car_df['Price'] = car_df['Price'].str.replace(r'\D', '', regex=True).astype('int64')
```

```python
#This is to test the results of the above cell.
car_df
```

```python
#6.)
car_df["Total Sales"] = car_df["Price"].cumsum()
car_df
```

```python
#7.)
car_df["Sale date"] = pd.date_range(
    start=pd.Timestamp.today().normalize(),
    periods=len(car_df),
    freq="D"
)
car_df
```

```python
#8.)
plt.figure()
plt.plot(car_df["Price"], car_df["Total Sales"])
plt.xlabel("Price")
plt.ylabel("Total Sales")
plt.title("Price vs total sales line plot")
plt.show()
```

```python
#9)
plt.figure(figsize=(10,5))
plt.scatter(car_df["Odometer (KM)"], car_df["Price"])
plt.xlabel("Odometer (KM)")
plt.ylabel("Price")
plt.title("Odometer vs Price scatter plot")
plt.show()
```

```python
#10.)
plt.figure(figsize=(12,5))
plt.bar(car_df["Make"], car_df["Odometer (KM)"])
plt.xlabel("Make")
plt.ylabel("Odometer (KM)")
plt.title("Make vs Odometer Bar Plot")
plt.show()
```

```python
#11.)
sns.histplot(car_df.Make, bins=10)
plt.title('Histogram of Car Make')
plt.show()
```

```python
#12.)
plt.figure(figsize=(14,12))

plt.subplot(2,2,1)
plt.plot(car_df["Price"], car_df["Total Sales"])
plt.xlabel("Price")
plt.ylabel("Total Sales")
plt.title("Price vs total sales line plot")

plt.subplot(2,2,2)
plt.scatter(car_df["Odometer (KM)"], car_df["Price"])
plt.xlabel("Odometer (KM)")
plt.ylabel("Price")
plt.title("Odometer vs Price scatter plot")

plt.subplot(2,2,3)
plt.bar(car_df["Make"], car_df["Odometer (KM)"])
plt.xlabel("Make")
plt.ylabel("Odometer (KM)")
plt.title("Make vs Odometer Bar Plot")

plt.subplot(2,2,4)
sns.histplot(car_df.Make, bins=10)
plt.title('Histogram of Car Make')

plt.show()
```

```python
#13.)
sns.boxplot(x='Colour', y='Total Sales', data=car_df)
plt.show()
```

```python
#14.)
sns.lmplot(x="Total Sales", y="Odometer (KM)", hue="Make", data=car_df, ci=None)
plt.show()
```

```python
#15.)
sns.pairplot(car_df)
plt.show()
```

```python
#16.)
car_subset = ['Total Sales', 'Price', 'Odometer (KM)']
car_pearson = car_df[car_subset]
correlation_matrix = car_pearson.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True)
plt.show()
#Price somewhat correlates with total sales, and Odometer does not correlate with Total sales.
```
